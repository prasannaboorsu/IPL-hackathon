### 10.To plot the most consistent batsmen among top 10 run getters, what plotting technique we use?

```python
a=pd.DataFrame(df.groupby (by=['batsman'])['batsman_runs'].sum())
a
```

```python
b=a.sort_values(by=['batsman_runs'] , ascending=False).head(10)

b
```

```python
sns.scatterplot (x = b.index, y = b['batsman_runs'], color ='c')
plt.title (" top 10 run getters, ")
plt.ylabel ("batsman_runs")
plt.xlabel ("batsman")
plt.xticks (rotation = 90)
plt.xticks (fontweight = 'bold')
```
